```python
class cocoDataSet_old1(data.Dataset):
    """
    加载coco数据集的类
    img_path:图片存放路径
    label_path:标签路径
    augment:是否需要增广
    """
    def __init__(self,img_path,label_path,augment,size_x,size_y):
        
        self.img_path = img_path
        self.label_path = label_path
        self.augment = augment
        self.size_x = size_x
        self.size_y = size_y
        
        img_list = [] 
        for _, _, files in os.walk(img_path): 
            img_list = [file[:-4] for file in files if file[-4:] == '.jpg'] 
        self.img_list = img_list 
        self.num_samples = len(self.img_list)

    def __getitem__(self,index):
        file_label = "%s%s.txt" %(self.label_path,self.img_list[index])
        file_img = "%s%s.jpg" %(self.img_path,self.img_list[index])
        with open(file_label,'r') as f:
            lines = f.readlines()
        f.close()
        classes = lines[0].replace('[',' ').replace(']',' ').replace(',',' ').split()
        classes = [int(i) for i in classes]
        boxes = lines[1].replace('[',' ').replace(']',' ').replace(',',' ').split()
        boxes = [float(i) for i in boxes]
        classes = np.array(classes)
        boxes = np.array(boxes).reshape(-1,4)
        img = cv2.imread(file_img)
        img,boxes = self.img_resize(img,boxes,self.size_x,self.size_y)
        img = img.transpose(2,0,1)
        
        return img,classes,boxes
    
    def img_resize(self,img,boxes,new_x,new_y):
        y,x,__ = img.shape
        scalex = float(new_x)/x
        scaley = float(new_y)/y
        img_trans = cv2.resize(img,(new_x,new_y))
        box_tool = np.array([scalex,scaley,scalex,scaley])
        box_trans = boxes * box_tool
        return img_trans,box_trans

    def __len__(self):
        
        return self.num_samples
        
```

```python
def img_flip(img,boxes):
    __,x,__ = img.shape
    boxes[:,0] = float(x) - boxes[:,0] - boxes[:,2]

    img_flip = cv2.flip(img,1)

    return img_flip,boxes
```

```python
class cocoDataSet(data.Dataset):
    """
    加载coco数据集的类
    img_path:图片存放路径
    label_path:标签路径
    augment:是否需要增广
    """
    def __init__(self,img_path,label_path,augment,size_x,size_y):
        
        self.img_path = img_path
        self.label_path = label_path
        self.augment = augment
        self.size_x = size_x
        self.size_y = size_y
        
        img_list = [] 
        for _, _, files in os.walk(img_path): 
            img_list = [file[:-4] for file in files if file[-4:] == '.jpg'] 
        self.img_list = img_list 
        self.num_samples = len(self.img_list)

    def __getitem__(self,index):
        file_label = "%s%s.txt" %(self.label_path,self.img_list[index])
        file_img = "%s%s.jpg" %(self.img_path,self.img_list[index])
        with open(file_label,'r') as f:
            lines = f.readlines()
        f.close()
        classes = lines[0].replace('[',' ').replace(']',' ').replace(',',' ').split()
        classes = [int(i) for i in classes]
        boxes = lines[1].replace('[',' ').replace(']',' ').replace(',',' ').split()
        boxes = [float(i) for i in boxes]
        classes = np.array(classes)
        boxes = np.array(boxes).reshape(-1,4)
        img = cv2.imread(file_img)
        img,boxes = self.img_resize(img,boxes,self.size_x,self.size_y)
        if self.augment:
            img,boxes = self.img_flip(img,boxes)
            img = self.img_blur(img)
        #此处需要处理标签和图像包括单位化 标签编码等
        img = img.transpose(2,0,1)
        img = img/255.
        img = torch.from_numpy(img).float()
        boxes[:,0]=(boxes[:,0]+boxes[:,2]/2)/self.size_x
        boxes[:,1]=(boxes[:,1]+boxes[:,3]/2)/self.size_y
        boxes[:,2]/=self.size_x
        boxes[:,3]/=self.size_y

        return img,classes,boxes
    
    def img_resize(self,img,boxes,new_x,new_y):
        
        y,x,__ = img.shape
        scalex = float(new_x)/x
        scaley = float(new_y)/y
        img_trans = cv2.resize(img,(new_x,new_y))
        box_tool = np.array([scalex,scaley,scalex,scaley])
        box_trans = boxes * box_tool
        return img_trans,box_trans
    
    def img_flip(self,img,boxes):
        
        if np.random.random() < 0.5:
            __,x,__ = img.shape
            boxes[:,0] = float(x) - boxes[:,0] - boxes[:,2]
            img= cv2.flip(img,1)
        return img,boxes

    def img_blur(self,img):
        if np.random.random() < 0.5:
            img = cv2.blur(img, (5, 5))
        return img
    #此处还可以添加亮度色调饱和度的增广
    
    def __len__(self):
        
        return self.num_samples
        
```
